# Modify improperly formatted data.¶

Fixing the data format for :
- gender
- customer_type
- age
- type_of_travel
- food_and_drink
- inflight_entertainment
- baggage_handling
- satisfaction.
- departure/arrival_time_convenient
- ease_of_online_booking

```python
for col in ['gender', 'customer_type', 'age', 'type_of_travel',
           'food_and_drink', 'inflight_entertainment', 'baggage_handling','ease_of_online_booking', 'departure/arrival_time_convenient','cleanliness','satisfaction']:
    print (col)
    unique = df_Business[col].unique()
    print('#.of unique values: ', len(unique), '\nList of unique values: ',unique, '\n*************\n')
```

```python
df_Business['gender'] = df_Business['gender'].replace(['F','M'],['Female', 'Male'])
df_Business['age'] = df_Business['age'].replace([180.0,185.0],[80.0,85.0])
df_Business['customer_type'] = df_Business['customer_type'].replace(['disloyal','Loyal'],
                                                                    ['Disloyal Customer','Loyal Customer'])
df_Business['type_of_travel'] = df_Business['type_of_travel'].replace(['Business','Business travel'],
                                                                      ['Business Travel','Business Travel'])
df_Business['food_and_drink'] = df_Business['food_and_drink'].replace(['Great'],['5'])
df_Business['inflight_entertainment'] = df_Business['inflight_entertainment'].replace([10.0],[5.0])
df_Business['baggage_handling'] = df_Business['baggage_handling'].replace(['A'],['5'])
df_Business['satisfaction'] = df_Business['satisfaction'].replace(['Dissatisfied'],['dissatisfied'])

```

```python
for col in ['gender', 'customer_type', 'age', 'type_of_travel',
           'food_and_drink', 'inflight_entertainment', 'baggage_handling','ease_of_online_booking','departure/arrival_time_convenient','cleanliness','satisfaction']:
    print (col)
    unique = df_Business[col].unique()
    print('#.of unique values: ', len(unique), '\nList of unique values: ',unique, '\n*************\n')
```

# Handle missing values

```python
df_Business.isnull().sum()
```

```python
mode_value_inflight = df_Business['inflight_wifi_service'].mode()
mode_value_inflight[0]
```

```python
df_Business['inflight_wifi_service'].fillna(mode_value_inflight[0], inplace=True)
```

```python
mode_value_da = df_Business['departure/arrival_time_convenient'].mode()
mode_value_da[0]
```

```python
df_Business['departure/arrival_time_convenient'].fillna(mode_value_da[0], inplace=True)
```

```python
mode_value_ease = df_Business['ease_of_online_booking'].mode()
mode_value_ease[0]
```

```python
df_Business['ease_of_online_booking'].fillna(mode_value_ease[0], inplace=True)
```

```python
mode_value_fd = df_Business['food_and_drink'].mode()
mode_value_fd[0]
```

```python
df_Business['food_and_drink'].fillna(mode_value_fd[0], inplace=True)
```

```python
mode_value_ie = df_Business['inflight_entertainment'].mode()
mode_value_ie[0]
```

```python
df_Business['inflight_entertainment'].fillna(mode_value_ie[0], inplace=True)
```

```python
mode_value_bh = df_Business['baggage_handling'].mode()
mode_value_bh[0]
```

```python
df_Business['baggage_handling'].fillna(mode_value_bh[0], inplace=True)
```

```python
mode_value_c = df_Business['cleanliness'].mode()
mode_value_c[0]
```

```python
df_Business['cleanliness'].fillna(mode_value_c[0], inplace=True)
```

```python
df_Business.isnull().sum()
```

Now we change other columns' null value into 'Unknown'

Now we replace rating 0 score into their mode values:

```python
df_Business['cleanliness'] = df_Business['cleanliness'].replace([0.0],[mode_value_c[0]])
df_Business['departure/arrival_time_convenient'] = df_Business['departure/arrival_time_convenient'].replace([0.0],[mode_value_da[0]])
df_Business['ease_of_online_booking'] = df_Business['ease_of_online_booking'].replace(['0','Unknown'],[mode_value_ease[0],mode_value_ease[0]])
df_Business['inflight_entertainment'] = df_Business['inflight_entertainment'].replace([0.0],[mode_value_ie[0]])
df_Business['food_and_drink'] = df_Business['food_and_drink'].replace(['0'],[mode_value_fd[0]])
```

```python
for col in ['gender', 'customer_type', 'age', 'type_of_travel',
           'food_and_drink', 'inflight_entertainment', 'baggage_handling','ease_of_online_booking','departure/arrival_time_convenient','cleanliness','satisfaction']:
    print (col)
    unique = df_Business[col].unique()
    print('#.of unique values: ', len(unique), '\nList of unique values: ',unique, '\n*************\n')
```

```python
df_Business['gender'].fillna('Unknown', inplace=True)
df_Business['customer_type'].fillna('Unknown', inplace=True)
df_Business['age'].fillna('Unknown', inplace=True)
df_Business['type_of_travel'].fillna('Unknown', inplace=True)
df_Business['flight_distance'].fillna('Unknown', inplace=True)
df_Business['departure_delay_in_minutes'].fillna('Unknown', inplace=True)
df_Business['arrival_delay_in_minutes'].fillna('Unknown', inplace=True)
df_Business['satisfaction'].fillna('Unknown', inplace=True)
```
